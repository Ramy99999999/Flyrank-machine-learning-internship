## Research Question

### Research Question

Which content pages should be reviewed first for refresh based on their
observed search performance and engagement signals?

### Decision

The project supports the decision of which pages should be prioritized
for review.

### Action

A content or SEO reviewer can inspect the highest-ranked pages and decide
whether they should be refreshed, expanded, rewritten, monitored, or left
unchanged.

### Unit of Analysis

One content page.

### Output

A ranked review queue that assigns each page a priority score together
with reason codes explaining why it was recommended.

## Quick Look at the Data

The starter dataset contains **30,000 content pages** described by
**44 features**, providing a solid foundation for exploring refresh
opportunities.

Of these pages, **16,262** are currently labelled with a **"down"**
trend direction. This suggests that a large portion of the dataset
contains pages that may require further investigation or review.

On average, each page receives **5,200.37 impressions** and
**37.07 sessions** over the 90-day period. These values indicate
meaningful variation in search visibility and user engagement,
making the dataset suitable for building a ranked review queue.

These observations support choosing the **Refresh / Content Opportunity
Scoring** lane because they show that the dataset contains measurable
signals that can help prioritize which pages should be reviewed first.
